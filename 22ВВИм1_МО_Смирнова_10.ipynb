```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import statistics as st
```

```python
df = pd.read_csv('/content/gt_2011.csv')
df.head(20)
```

```python
df.shape
```

```python
GTEP = df['GTEP']
TEY = df['TEY']
fig, ax = plt.subplots()
ax.scatter(GTEP, TEY, c='blue', alpha=0.3, edgecolors='none')

ax.grid(True)
plt.xlabel('GTEP')
plt.ylabel('TEY')
plt.show()
```

```python
print("Математическое ожидание: ")
print(st.mean(df['GTEP']))

print("\nСредневзвешенное значение: ")
print(np.average(df['GTEP']))

print("Медиана: ", st.median(df['GTEP']))
print("Медиана меньшее среднее значение: ", st.median_low(df['GTEP']))
print("Медиана большее среднее значение: ", st.median_high(df['GTEP']))

print('Значение моды: ', st.mode(df['GTEP']))

print("Минимум: ", np.min(df['GTEP']))
print("Максимум: ", np.max(df['GTEP']))
```

```python
print("Математическое ожидание: ")
print(st.mean(df['TEY']))

print("\nСредневзвешенное значение: ")
print(np.average(df['TEY']))

print("Медиана: ", st.median(df['TEY']))
print("Медиана меньшее среднее значение: ", st.median_low(df['GTEP']))
print("Медиана большее среднее значение: ", st.median_high(df['GTEP']))

print('Значение моды: ', st.mode(df['TEY']))

print("Минимум: ", np.min(df['TEY']))
print("Максимум: ", np.max(df['TEY']))
```

```python
print("Математическое ожидание: ")
print(st.mean(df['AT']))

print("\nСредневзвешенное значение: ")
print(np.average(df['AT']))

print("Медиана: ", st.median(df['AT']))
print("Медиана меньшее среднее значение: ", st.median_low(df['AT']))
print("Медиана большее среднее значение: ", st.median_high(df['AT']))

print('Значение моды: ', st.mode(df['AT']))

print("Минимум: ", np.min(df['AT']))
print("Максимум: ", np.max(df['AT']))
```

```python
# Метрики оценки вариативности данных

print("Дисперсия: ")
print(st.pvariance(df['GTEP']))

print("\nСтандартное отклонение: ")
print(st.pstdev(df['GTEP']))

print("\nКвантиль: ")
print(np.percentile(df['GTEP'], 5))
print(np.percentile(df['GTEP'], 95))
print(np.percentile(df['GTEP'], [25, 50, 75]))

print("\nРазмах: ")
GTEP_range = max(df['GTEP']) - min(df['GTEP'])
print(GTEP_range)

x = [i for i in range(0, len(df['GTEP']))]
y = df.loc[:, 'GTEP']
plt.xlabel('Номер')
plt.ylabel('Уровень давления выхлопных газов газовой турбины')
plt.title('Точечный график зависимости двух выборок x и y')
plt.grid(True)
plt.scatter(x, y, marker='.')
plt.show()
```

```python
# Метрики оценки корреляции или взаимозависимости
GTEP = df['GTEP']
TEY = df['TEY']
plt.xlabel('выборка GTEP')
plt.ylabel('выборка TEY')
plt.title('Точечный график зависимости двух выборок x и y')
plt.grid(True)
plt.scatter(GTEP, TEY, marker='.')
```

```python
from scipy.stats import pearsonr, spearmanr, kendalltau, norm
```

```python
corr, p_value = pearsonr(GTEP, TEY)
plt.scatter(GTEP, TEY, marker='.')
print('Коэффициент корреляции Пирсона:', corr)
```

```python
cor, p_value = sts.spearmanr(GTEP, TEY)
print('Коэффициент корреляции Спирмена:', cor)
```

```python
cor, p_value = sts.kendalltau(GTEP, TEY)
print('Коэффициент корреляции Тау Кендалла:', cor, p_value)
```
