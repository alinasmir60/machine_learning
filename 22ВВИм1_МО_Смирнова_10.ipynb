```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import statistics as st
```

```python
df = pd.read_csv('/content/gender_classification_v7.csv')
df.head(20)
```

```python
df.shape
```

```python
df.gender.unique()
```

```python
Gender = df['gender']
forehead_width_cm = df['forehead_width_cm']
fig, ax = plt.subplots()
ax.scatter(Gender, forehead_width_cm, c='blue', alpha=0.3, edgecolors='none')

ax.grid(True)
plt.xlabel('gender')
plt.ylabel('forehead_width_cm')
plt.show()
```

```python
print("Математическое ожидание: ")
print(st.mean(df['forehead_width_cm']))

print("\nСредневзвешенное значение: ")
print(np.average(df['forehead_width_cm']))

print("Медиана: ", st.median(df['forehead_width_cm']))
print("Медиана меньшее среднее значение: ", st.median_low(df['forehead_width_cm']))
print("Медиана большее среднее значение: ", st.median_high(df['forehead_width_cm']))

print('Значение моды: ', st.mode(df['forehead_width_cm']))

print("Минимум: ", np.min(df['forehead_width_cm']))
print("Максимум: ", np.max(df['forehead_width_cm']))
```

```python
print("Математическое ожидание: ")
print(st.mean(df['forehead_height_cm']))

print("\nСредневзвешенное значение: ")
print(np.average(df['forehead_height_cm']))

print("Медиана: ", st.median(df['forehead_height_cm']))
print("Медиана меньшее среднее значение: ", st.median_low(df['forehead_height_cm']))
print("Медиана большее среднее значение: ", st.median_high(df['forehead_height_cm']))

print('Значение моды: ', st.mode(df['forehead_height_cm']))

print("Минимум: ", np.min(df['forehead_height_cm']))
print("Максимум: ", np.max(df['forehead_height_cm']))
```

```python
# Метрики оценки вариативности данных

print("Дисперсия: ")
print(st.pvariance(df['forehead_width_cm']))

print("\nСтандартное отклонение: ")
print(st.pstdev(df['forehead_width_cm']))

print("\nКвантиль: ")
print(np.percentile(df['forehead_width_cm'], 5))
print(np.percentile(df['forehead_width_cm'], 95))
print(np.percentile(df['forehead_width_cm'], [25, 50, 75]))

print("\nРазмах: ")
GTEP_range = max(df['forehead_width_cm']) - min(df['forehead_width_cm'])
print(GTEP_range)

x = [i for i in range(0, len(df['forehead_width_cm']))]
y = df.loc[:, 'forehead_width_cm']
plt.xlabel('Номер')
plt.ylabel('Ширина лба')
plt.title('Точечный график зависимости двух выборок x и y')
plt.grid(True)
plt.scatter(x, y, marker='.')
plt.show()
```

```python
print("Дисперсия: ")
print(st.pvariance(df['forehead_height_cm']))

print("\nСтандартное отклонение: ")
print(st.pstdev(df['forehead_height_cm']))

print("\nКвантиль: ")
print(np.percentile(df['forehead_height_cm'], 5))
print(np.percentile(df['forehead_height_cm'], 95))
print(np.percentile(df['forehead_height_cm'], [25, 50, 75]))

print("\nРазмах: ")
TEY_range = max(df['forehead_height_cm']) - min(df['forehead_height_cm'])
print(TEY_range)

x = [i for i in range(0, len(df['forehead_height_cm']))]
y = df.loc[:, 'forehead_height_cm']
plt.xlabel('Номер')
plt.ylabel('Высота лба')
plt.title('Точечный график зависимости двух выборок x и y')
plt.grid(True)
plt.scatter(x, y, marker='.')
plt.show()
```

```python
# Метрики оценки корреляции или взаимозависимости
forehead_height_cm = df['forehead_height_cm']
forehead_width_cm = df['forehead_width_cm']
plt.xlabel('выборка forehead_height_cm')
plt.ylabel('выборка forehead_width_cm')
plt.title('Точечный график зависимости двух выборок x и y')
plt.grid(True)
plt.scatter(forehead_height_cm, forehead_width_cm, marker='.')
```

```python
from scipy.stats import pearsonr, spearmanr, kendalltau, norm
```

```python
corr, p_value = pearsonr(forehead_height_cm, forehead_width_cm)
plt.scatter(forehead_height_cm, forehead_width_cm, marker='.')
print('Коэффициент корреляции Пирсона:', corr)
```

```python
cor, p_value = sts.spearmanr(forehead_height_cm, forehead_width_cm)
print('Коэффициент корреляции Спирмена:', cor)
```

```python
cor, p_value = sts.kendalltau(forehead_height_cm, forehead_width_cm)
print('Коэффициент корреляции Тау Кендалла:', cor, p_value)
```
